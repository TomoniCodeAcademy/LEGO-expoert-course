# サーボモータを単体でうごかします

サーボモータを使う上で、LEGOカーを動かす場合はサーボモータを２つペアで動かします。サーボでアームを動かす場合は、サーボモータを単体で使います。ここでは、単体で動かす方法を説明します。

```python
from spike import Motor

motor = Motor('A')                # Connect the motor to port A
motor.start(speed=20)             # speed (-100 ～ 100)
```

モータの回転を止めるには 関数stopを使います。

```python
from spike import Motor

motor = Motor('A')                # Connect the motor to port A
motor.stop()
```

指定した角度まで回す、指定した回転数を回すこともできます。

# サーボモータを指定した角度まで動かす

サーボモータを指定した角度まで動かしたい場合は、関数 run_for_degreesを使います<br>
下のプログラムは、サーボモータを時計まわりに90度まで回します(スピードはゆっくり(20%))

```python
from spike import Motor

# モーターを初期化します。
motor = Motor('A')
motor.run_for_degrees(90, speed=20)    # 時計回りに90度まで回転します
```
