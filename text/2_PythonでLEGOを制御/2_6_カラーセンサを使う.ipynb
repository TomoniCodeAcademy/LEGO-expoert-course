# カラーセンサの値を取得する（色の名前を取得）

カラーセンサの値を取得するには、カラーセンサを初期化した後、関数get_colorを使います<br>
file:  SENSOR_distance_00.xxxx

```python
from spike import ColorSensor

# カラーセンサーを初期化します
color_sensor = ColorSensor('C')

# 色を測定します
color = color_sensor.get_color()
print(color)
```

一度だけの実行の場合、正確に色を読み取れない場合があるので、動作テストする時は下記のようなループを作って何度も読み込んでみてください。

```python
from spike import ColorSensor
import time

# カラーセンサーを初期化します
color_sensor = ColorSensor('C')

# 色を測定します
while True:
    color = color_sensor.get_color()
    print(color)
    time.sleep(1)
```

# カラーセンサの値を取得する(R,G,Bの３原色)

色の名前ではなく、RGBの３原色で取得することもできます。<br>
３原色を取得したい場合、関数get_rgb_intensityを使います。<br>
下記プログラムの場合、変数colorにはR,G,Bそれぞれの色の強さが入ります。

```python
from spike import ColorSensor
import time

# カラーセンサーを初期化します
color_sensor = ColorSensor('C')
while True:
    color=color_sensor.get_rgb_intensity()
    print(color)
    time.sleep(1)
```
