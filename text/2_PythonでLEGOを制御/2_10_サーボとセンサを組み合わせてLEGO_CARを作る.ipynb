# カラーセンサとサーバを組み合わせたプログラム

カラーセンサとサーボを組み合わせて、ライントレースするLEGO　CARを作りましょう。プログラムは以下の通りです。LEGO CARが黒線から外れているかどうかを判断します。もし外れている時は車を右に曲げます。もしラインの上なら、車を左に曲げます。
これを繰り返すことで、黒いラインに沿ってLEGO CARを走らせることが可能になります。
下記はLEGO CARの外観です。<br>

ソースコードは以下の通りです、カラーセンサから得られる明るさを使って、LEGO Carが白いパネルの上か、ライン上であるかを判断します。もし明るい場合は、白いパネルの上であると判断し、ハンドルを右に曲げます。もし暗い場合はライン上であると判断し、ハンドルを左に曲げます。この操作を繰り返すことで、LEGO CARはラインに沿って走ることができます。

```python
from spike import MotorPair
from spike import ColorSensor

# モータを初期化します
motor_pair = MotorPair('A','B')
motor_pair.set_default_speed(10)

# カラーセンサーを初期化します
color_sensor = ColorSensor('C')

while True:   # ずっと繰り返す
    # カラーセンサから明るさを取得
    light=color_sensor.get_reflected_light()
    print(light)
    if light>70:
        # もし白色の上ならLEGO CARを右に曲げる
        print('>>>>>>>>>>>>>>>>>>>>')
        motor_pair.start(steering = 50)
    else:
        # そうでないならLEGO CARを左に曲げる
        print('<<<<<<<<<<<<<<<<<<<<<<')
        motor_pair.start(steering = -50)
```
