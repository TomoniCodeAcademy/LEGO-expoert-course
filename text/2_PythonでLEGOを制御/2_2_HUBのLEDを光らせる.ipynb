# イメージを表示させる

ライトマトリックスにイメージを表示するにはshow_image関数を使います<br>
下のプログラムは、'HAPPY'というイメージを表示します<br>
file: LED_matrix_00.llsp

```python
from spike import PrimiHub

hub = PrimeHub()
hub.light_matrix.show_image('HAPPY')
```

ライトマトリックスを消そう<br>
関数off()を使うとLEDが消えます

```python
from spike import PrimiHub
import time

hub = PrimeHub()
hub.light_matrix.show_image('HAPPY')
time.sleep(3)                         # ３秒待つ
hub.light_matrix.off()                # 表示を消して終わる
```


下のプログラムは全てのイメージを順番に表示します


```python
from spike import PrimeHub
import time

IMAGE_LIST = ('SMILE', 'ANGRY', 'ASLEEP', 'CONFUSED', 'FABULOUS', 'HAPPY', 'MEH',
'SAD', 'SILLY', 'SURPRISED','ARROW_N', 'ARROW_S', 'ARROW_E', 'ARROW_W', 'ARROW_NE',
'ARROW_NW', 'ARROW_SE', 'ARROW_SW','CLOCK1', 'CLOCK2', 'CLOCK3', 'CLOCK4', 'CLOCK5',
'CLOCK6', 'CLOCK7', 'CLOCK8', 'CLOCK9', 'CLOCK10', 'CLOCK11', 'CLOCK12','BUTTERFLY',
'COW', 'PACMAN', 'GHOST', 'RABBIT', 'SNAKE', 'DUCK', 'GIRAFFE', 'TORTOISE','DIAMOND',
'DIAMOND_SMALL', 'HEART', 'HEART_SMALL', 'SQUARE', 'SQUARE_SMALL', 'TRIANGLE',
'TRIANGLE_LEFT', 'CHESSBOARD', 'STICKFIGURE', 'TARGET', 'XMAS', 'MUSIC_CROTCHET',
'MUSIC_QUAVER', 'MUSIC_QUAVERS', 'HOUSE', 'PITCHFORK', 'ROLLERSKATE', 'SKULL',
'SWORD', 'TSHIRT', 'UMBRELLA', 'YES', 'NO')


hub = PrimeHub()
for image_name in IMAGE_LIST:                                # IMAGEのリストから一つずつ取り出す
    print(image_name)
    hub.light_matrix.show_image(image_name)                  # HUBのライトマトリックスに表示する
    time.sleep(1)

hub.light_matrix.show_image('HAPPY')
```

# 文字を表示する




ライトマトリクスに数字や文字を表示するにはwrite関数を使います<br>
file: LED_matrix_1.llsp

```python
from spike import PrimiHub

hub = PrimeHub()
hub.light_matrix.write('1')
```
