# カメラの使い方

AIカメラには機能を選択するダイアルと、対象物を学習するためのボタンがあります。
AIカメラの使い方として、①機能選択ダイアルを使って使いたい機能を選択、②認識したい対象にカメラを向けて、学習ボタンを押してカメラに覚えさせる、③学習済みのカメラを使って対象を認識する、の流れとなります。具体的な操作は以下に説明します。

# AIカメラで顔を認識する

あらかじめ顔を覚えさせることで、カメラに映る相手が知っている人かどうかを判断することができます(相手が誰かをカメラに覚えさせる操作を学習と呼びます)。複数の顔を覚えることができますが、プログラムを簡単にするため、学習する顔は一人だけとします。

まず左側のダイアルを使って機能を選択します。顔認識を行いたいので、Face Recognitionが表示されるまで、ダイアルを左側に倒します。Face Recognitionが表示されると、顔認識モードになります。<br>
<img src="https://github.com/TomoniCodeAcademy/LEGO-expoert-course/blob/main/img/pictures/HuskyLens_select_faceRecognition.png?raw=true" width=400>

顔認識モードに切り替えた時、画面に十字のマークが表示されていると学習モードになっています。十字マークが出ていると人の顔の覚えさせることができます。覚えたい人の顔にカメラを向けてボタンを押します。この操作により顔の学習が行われます。学習が済むと、顔の位置にFace：ID１と表示されます。<br>
<img src="https://github.com/TomoniCodeAcademy/LEGO-expoert-course/blob/main/img/pictures/HuskyLens_FaceIdentified.png?raw=true" width=400>

学習してない人の顔はFaceとだけ表示されます。これによりカメラを向けた相手が、知っている人(Face:ID1)か、知らない人(Face)か見分けることができます。<br>
<img src="https://github.com/TomoniCodeAcademy/LEGO-expoert-course/blob/main/img/pictures/HuskyLens_Face_notIdentified.png?raw=true" width=400>

顔認識モード(Face Recognition)に切り替えた時、画面に十字のマークが表示されず、Faceと表示される場合、カメラは学習済みで、認識モードで動いています。新しい顔を覚えさせるために、まず覚えた顔を忘れさせる必要があります。
忘れるためにまず右側のボタンを一度押します。すると、「忘れるには一度押して」(Click again to forget!)と表示されます。<br>
<img src="https://github.com/TomoniCodeAcademy/LEGO-expoert-course/blob/main/img/pictures/HuskyLens_forget.png?raw=true" width=400>

もう一度ボタンを押すと以前に覚えた顔は忘れて、学習モードになります。学習モードになったら、上記の手順に沿って、顔を覚えさせます。このようにAIカメラにあらかじめ人の顔を学習させます。カメラが映している範囲に学習した人の顔が入ってくると、AIカメラは知ってる人と判断します(Face:ID1)。

顔認識機能とLEGOのサーボモータを組み合わせることで、家の人が帰ってきたら自動で開くドアをつくることができます（家の人以外は開けません）。

## 演習
- AIカメラを顔認識モードに切り替えてください
- 友達を顔を学習させてみましょう。
- 学習した顔が正しく認識されるか確認してください。正しく認識されるとは、Face:ID1 と1番の数字が振られることを確認してください。
- 学習していない人を映しても、Faceとだけ表示されることを確認してください

# AIカメラで色を認識する




AIカメラには特定の色を覚えさせることで、学習した色がカメラの前にあるかどうかを判断することができます。また、学習した色がカメラの前にあると、画像内でX,Yの位置も特定してくれます。色認識機能とLEGOカーを組み合わせることで、色のついたピンポン玉を追いかけるLEGOカーをつくることができます。

色認識を行うには、Color Recognitionが表示されるまで、ダイアルを左側か右側に倒します。Color Recognitionが表示されると、色認識モードになります。

<img src="https://github.com/TomoniCodeAcademy/LEGO-expoert-course/blob/main/img/pictures/HuskyLens_ColorIRecognition.png?raw=true" width=400>

顔認識と同じように学習させたい色にカメラの十字マークを向けて右側のボタンを押します。カメラは指定された色を学習します。
学習が行われるとカメラが映している範囲に同じ色の物体が入ってくると、AIカメラはその物体を見つけ出すことができます。

<img src="https://github.com/TomoniCodeAcademy/LEGO-expoert-course/blob/main/img/pictures/HuskyLens_ColorIRecognition_learned_object.png?raw=true" width=400>

見つけ出した場合、画面内のどこに物体があるかを教えてくれます。



## 演習
- AIカメラを色認識モードに切り替えてください
- 学習させたい色を決めて学習してみてください
- 学習した色が正しく認識されるか確認してください。正しく認識されるとは、Color:ID1 と1番の数字が振られることを確認してください。
- 学習していない色には反応しないことも確認しましょう
