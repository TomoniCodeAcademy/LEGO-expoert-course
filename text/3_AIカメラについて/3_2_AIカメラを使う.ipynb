# AIカメラを使う（顔認識）

AIカメラにあらかじめ人の顔を学習させます（山田君の顔）。カメラが映している範囲に学習した人（人の顔の人）が入ってくると、AIカメラは人を見つけてくれます。見つけ出した場合、画面内のどこに人がいるかも教えてくれます。

顔認識機能とLEGOのサーボモータを組み合わせることで、自動で鍵が開くドア等をつくることができます。

# AIカメラを使う（色認識）

AIカメラにあらかじめ色を学習させます（オレンジ色のピンポン玉）。カメラが映している範囲に同じ色の物体が入ってくると、AIカメラはその物体を見つけ出すことができます。見つけ出した場合、画面内のどこに物体があるかを教えてくれます。

色認識機能とLEGOカーを組み合わせることで、オレンジ色のピンポン玉を追いかけるLEGOカーをつくることができます。
