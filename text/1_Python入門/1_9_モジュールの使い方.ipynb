# Pythonのモジュールと、importについて

Pythonはシンプルな言語仕様であり、最低限の機能に絞って作られています。何か複雑な処理をしたい場合は、機能を取り込んで機能を追加します。
例えば、、家を建てようと思った時に、道具箱を借りてきて、のこぎりや金づちを手に入れるようなものです。

Pythonにはこのような道具箱がたくさん用意されていて、必要になるたびに、道具箱を借りてきて機能アップします。道具箱のことをPythonでもモジュール(module)と呼びます。
用意されている主なモジュールとしては以下があります。

*  math     算数の計算用モジュール
* random   乱数を作るモジュール（ゲーム等で活躍）
* datetime 日付を計算するモジュール（今日の日付や時刻を知る時につかう）

来月ぐらいに出てきますが、PythonからLEGOを制御する時、LEGO を制御するためのモジュールを呼び出して使います。主なモジュールは以下です。

* spike    LEGO SPIKEを使うためのモジュール
* HuskyLes  AIカメラを使うためのモジュール

モジュールの機能を呼び出すにはimport文を使います。例えば、サーボモータを使う場合は以下のように書きます。

```python
from spike import Motor
```

AIカメラ用のモジュールを使いたい場合は以下の様に書きます。

```python
from huskylens_lib import HuskyLens
```

今後の授業では、あらかじめサンプルソースの行頭にimport 文を付けていますので、生徒の皆さんは直接書き足すことはありません。import文を使って機能を取り込んでいるのだなとご理解ください。
