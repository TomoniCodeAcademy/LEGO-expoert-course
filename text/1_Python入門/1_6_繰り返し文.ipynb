# 指定された回数繰り返す

回数を指定して繰り返す場合はforループを使います。以下のサンプルは5回繰り返す例です。動かしてみましょう

```python
# 繰り返しサンプル
for _ in range(5):
    print('hello')
```

上記プログラムにおいて、for _ in range(数字):  は決まり文句です。range(数字)に設定する数字を変えることで、繰り返し回数を変えることができます。 繰り返ししたい範囲(for文のbody部と呼ばれます)はソースコードの段下げで指示します（この説明はややこしいので先生の補足を聞いてください）

# 無限に繰り返す

LEGOプログラミングにおいて、ずっと繰り返すブロックがありました。<br>
<img src='https://github.com/TomoniCodeAcademy/LEGO-expoert-course/blob/main/img/document_parts/mugen_loop_block.png?raw=true' width=150>

Pythonにおいて無限に繰り返す場合は、whileループを使います。以下の例では無限にprintを実行します。<br>
いつまでも走り続ける（暴走する）ので、■マークを押して実行を止めてください。

```python
while True:
    print('hello')
```

無限に繰り返す場合、　while True:  と書くのもPythonでは決まり文句です。覚えましょう。（忘れたらこの章を思い出してください）

**発展:**
while True:   について少しだけ説明します。  while文は、　while ***<何とか>*** : と書きます。　***<何とか>***の部分は条件式を書きます。条件式を計算した結果がTrueの場合while文のbody部を繰り返します。while True:  と書くといつまでもTrueなので、body部はいつまでも実行されることになります。 ***＜何とか＞***　の部分の書き方は＃＃を見てください。

# 条件が成立する間、繰り返す

while文と条件式と組み合わせることで条件が成立している間だけ繰り返すことができます。<br>
LEGO ブロックプログラミングでも、以下のように条件が成立するまで繰り返すブロックがありました。<br>
<img src="https://github.com/TomoniCodeAcademy/LEGO-expoert-course/blob/main/img/document_parts/until_loop.png?raw=true" width=150><br>
LEGOブロックでは、条件が成立するまでくりかえしました。正確に言うと、条件が成立するとループが終了しました。pythonの場合は、条件が成立するあいだ、処理を繰り返します。
<br>while に続けてチェックする条件を書きます。<br> 例は以下です。<br>
条件式の詳しい書き方は判断の章を見てください。

```python
x = 0            # 変数x を 0に設定
while x < 10:    # x が10より小さい間はループを続ける
  print(x)
  x += 1         # xを１増やす
```

ご参考：上記プログラムは以下のプログラムと同じ動きとなります

```python
for x in range(10):
  print(x)
```

# 発展：リストから値を取り出しながら繰り返す

Pythonではリストから値を取り出しながら繰り返すことができます。この書き方を使うことで、簡潔で分かりやすいコードを書くことができます。<br>
決まり文句としては、<br>
  for <変数> in  <リスト> : <br>
の形式です。

以下の例では、リスト(10, 20, 100)から値を取り出しながらループします。リストから値が一つずつ取り出しながら実行されるのを確認してください。

```python
for  x  in  (10,20,30):
    print(x)
```

下の例は、文字列を順番に取り出しながらループします

```python
for x in ('hello','good bye','see you'):
    print(x)
```
