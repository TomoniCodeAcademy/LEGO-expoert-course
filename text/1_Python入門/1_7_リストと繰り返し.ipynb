# リスト型

いろんな値をまとめて扱う方法として、ブロックプログラミングでは配列を使うことできました。ブロックの形としては以下です。


Pythonでも、値をまとめて使う方法が用意されています。リスト型と呼ばれます。書き方は以下のように、値をカンマ(",")で並べて、カッコ("[" , "]")で囲みます。リストの中には、どんな値でも入れることができます。


```python
my_list = [1,2,3,4,5]
my_friends = ['こうさく', 'ジョージ', '山田']
```

**発展：**
文法上、異なる型の値を入れることはできますが、バグになりやすいので、基本的には同じ型の値だけをリストに入れるのが望ましいです。

# リスト型の操作

リスト型に保存されている値の取り出しは、以下のように取り出したい位置を指定します。２番目の'ジョージ'を取り出す場合、1を指定します。最初の場所は０番ですので注意してください。

```python
my_friends = ['こうさく', 'ジョージ', '山田']
my_friends[1]
```

リスト型の指定の場所に値を設定するには以下のようにします。

**発展：**
番地が０番から始まることを０始まりと呼びます。１番から始まるのは１始まりと呼びます。Pythonは0始まりの言語です。プログラミング言語によって、０始まりで指定する約束の場合と１始まりで指定する約束の場合があります。間違えると違う値を取ってきますので注意してください。プログラミングのミスが発生しやすい所です。

# リストから値を取り出しながらくりかえす

Pythonではリストから値を取り出しながら繰り返し実行することができます。この書き方を使うことで、簡潔で分かりやすいコードを書くことができます。 決まり文句としては、 for <変数> in <リスト> : の形式です。

以下の例では、リスト(10, 20, 100)から値を取り出しながらループします。リストから値が一つずつ取り出しながら実行されるのを確認してください。

```python
for  x  in  (10,20,30):
    print(x)
```

下の例は、文字列をリストから順番に取り出しながらループします

```python
for x in ('hello','good bye','see you'):
    print(x)
```

for文に与えるリストを変数で指定することもできます。以下の例は、友人の名前の入った変数, my_friends を指定して、for 文を実行する例です。友達の名前を順番に出力します。

```python
my_friends = ['こうさく', 'ジョージ', '山田']
for friend in my_friends:
    print(friend)
```
