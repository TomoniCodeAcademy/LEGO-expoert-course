# リスト型

いろんな値をまとめて扱う方法として、ブロックプログラミングではリストを使うことできました。ブロックの形は以下です。リストを作るボタンを押してリストを作成しましたね。<br>
<img src="https://github.com/TomoniCodeAcademy/LEGO-expoert-course/blob/main/img/document_parts/list_blocks.png?raw=true" width=300>


Pythonでも、値をまとめて使う方法が用意されています。リスト型と呼ばれます。書き方は以下のように、値をカンマ(",")で並べて、カッコ("[" , "]")で囲みます。リストの中には、どんな値でも入れることができます。


```python
my_friends = ['こうさく', 'ジョージ', '山田太郎']
```

発展：
文法上、異なる型の値を入れることはできますが、バグになりやすいので、基本的には同じ型の値だけをリストに入れるのが望ましいです。int型だけ入れる、あるいは、文字列型だけ入れるというように、同じ型の値だけを入れましょう。

# リスト型に対する値の操作

リストに入れた値を使うには、リストから値を取り出す必要があります。ブロックには、以下のような、値を取り出すブロック、値を設定するブロックがありました。<br>
<img src="https://github.com/TomoniCodeAcademy/LEGO-expoert-course/blob/main/img/document_parts/list_blocks_access.png?raw=true" width=400>

Pythonにおいてリスト型に保存されている値の取り出しは、以下のように取り出したい位置をカッコ('[',']')で指定します。リスト型の値が入ってる変数 my_friendsに対して、２番目の'ジョージ'を取り出す場合、1を指定して、 <br>
my_friends[1]<br>
と書きます。こうすると、Pythonは2番目の値を取り出してくれます。
最初の場所は０番ですので注意してください。

```python
my_friends = ['こうさく', 'ジョージ', '山田']
my_friends[1]
```

リスト型の特定の場所に値を設定するには以下のように位置を指定してイコール('=')を使って値を設定します。

```python
my_friends = ['こうさく', 'ジョージ', '山田']
my_friends[1] = 'トーマス'
```

**発展：**
番地が０番から始まることを０始まりと呼びます。１番から始まるのは１始まりと呼びます。Pythonは0始まりの言語です。LEGOのブロックプログラミングは1始まりでした。このようにプログラミング言語によって、０始まりで指定する約束の場合と１始まりで指定する約束の場合があります。間違えると違う値を取ってきますので注意してください。プログラミングのミスが発生しやすい所です。

# リストから値を取り出しながらくりかえす

Pythonではリストから値を取り出しながら繰り返し実行することができます。この書き方を使うことで、簡潔で分かりやすいコードを書くことができます。 決まり文句としては、 for <変数> in <リスト> : の形式です。

以下の例では、リスト(10, 20, 100)から値を取り出しながらループします。リストから値が一つずつ取り出しながら実行されるのを確認してください。

```python
for  x  in  (10,20,30):
    print(x)
```

下の例は、文字列をリストから順番に取り出しながらループします

```python
for x in ('hello','good bye','see you'):
    print(x)
```

for文に与えるリストを変数で指定することもできます。以下の例は、友人の名前の入った変数, my_friends を指定して、for 文を実行する例です。友達の名前を順番に出力します。

```python
my_friends = ['こうさく', 'ジョージ', '山田']
for friend in my_friends:
    print(friend)
```
