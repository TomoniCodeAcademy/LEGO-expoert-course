このように処理をまとめた新しいブロックを作ることがPythonでもできます。Pythonではブロックはありませんので、関数を使います。上記の新しいブロックの定義と同じ動きをPythonで書くと以下となります。(Pythonでは日本語が使えないので英語で書きます)

```python
def zigzag(distance):
    lego.motor_pair('left', distance)
    lego.motor_pair('straight', distance)
    lego.motor_pair('right', distance)
```

### 関数を定義する方法を説明します

① 関数定義は、まずdef と書きます(これは決まっています)。

② 次に関数名を書きます。関数の名前はなんでもいいのですが、今回、ジグザグ前進を英語にして、関数名はzigzagとしています。

③ 関数名に続けてカッコ() を書きます。関数を実行する際、値を指定する場合にカッコの中に変数を書きます。値を指定しない場合は変数は不要です。
関数zigzagを使う時、進む距離を指定して使います。カッコの中には指定された距離を受け取る変数を書きます（引数（ひきすう）と呼ばれます）。変数の名前はなんでもいいのですが、距離を表す英語のdistanceとしています。

④ カッコの後ろにコロン : をつけてください。これも決まりです。コロンを書くのを忘れてよくエラーになります。

⑤ 次に、zigzag関数では何をするのか？を書きます。関数zigzagでは、左曲がり、直進、右曲がりを行いますので、LEGOが用意する関数を書きます。この時、先頭は空白4文字空けてください。(インデントと呼ばれます)。インデントは空白4文字ではく3文字や2文字でもエラーにはなりませんが、みんなが守るルールとして、空白4文字を使うことが決まりとなっています。空白は半角を使ってください、全角（漢字）の空白を使うとエラーになります。

インデントを設定することで、Pythonに対して、zigzagの関数定義がどこまでかを示しています（関数本体、あるいは、body部と呼ばれます）。もう少し詳しく説明すると、Pythonとのやりとりにおいて、 : を書くことで、関数の本体の開始を表し、インデントが続く範囲が本体となります。

Pythonは、関数zigzagの本体が、左曲がり、直進、右曲がりであると分かります。インデントの位置をいろいろ混ぜるとエラーになりますので、インデントの位置は揃えるようにしてください。これはPythonを使ってプログラミングする際に一番多く発生するエラーと思います。

作られた関数を呼び出すのは以下です。距離10を指定して、関数zigzagを呼び出しています。

```python
zigzag(10)
```

簡単な関数を作ってみましょう。昨日食べた物を人に伝える関数を作ってみましょう。

関数名はbangohan、引数の変数は tabetamono としています。

```python
def  bangohan(tabetamono):
    print('昨日の夜は')
    print(tabetamono)
    print('を食べました')
```

作ったbangohan関数を呼び出してみましょう。カレーを食べたので、カレーを指定して関数を呼び出しています。

```python
bangohan('カレー')
```

エラーの原因と修正方法

- **\#error_01の場合**
  - エラーの原因：関数定義の行にコロン( : )が書かれていません
  - 修正方法：関数定義の行にコロン( : )を追加してください
- **\#error_02の場合**
  - エラーの原因：関数(hello)の本体のインデント(行頭の空白)が揃っていません
  - 修正方法：print('bye')の行頭の空白を４文字に修正してください
- **\#error_03の場合**
  - エラーの原因：関数(hello)の本体のインデント(行頭の空白)がありません
  - 修正方法：print('bye')の行頭に空白を４文字を追加してください
- **\#error_04の場合**
  - エラーの原因：ソースコード中に全角文字が使われています
  - 修正方法：print('bye')の行頭の全角文字を消して半角の空白を４文字を追加してください
