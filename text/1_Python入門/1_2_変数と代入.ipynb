Pythonも変数を使うことができます。Pythonで変数を使うには、使いたい変数名を決めて何か値を入れます。以下に新しい変数　my_nameを使う例を示します。

```python
my_name = 'ToMo-Ni Taro'     # 変数を初期化する（変数を作り、値を設定する）
```

上記の例の様に、変数に何か値を入れるには、イコール(=)を１つ使います。慣れないと何をしているのか分かりにくいと思いますが、変数名 =  値　と書くことで、新しい変数が一つ作られ、指定した値が設定されます。変数に値を入れることを変数に値を設定すると言います。プログラム中で、変数の値を何度も設定しますが、一番初めに変数を値をすることを変数の初期化と言います。

一度変数を作ると、print関数で変数名を使って出力することができます。以下は変数名 your_nameを作成して出力する例です。

```python
your_name = 'ToMo-Ni Hanako'   # 変数を初期化する
print(your_name)               # 変数 your_nameの値を出力する
```

演習：変数を作って値を設定してください。

変数には数値や文字列を入れることができます。変数の値を使って演算することができます。以下の例は変数の値を使って面積を計算する例です。計算した結果を新しい変数、areaに設定しています。

```python
tate = 20   #  たて　が　　20cm
yoko = 30   # 　横が  30cm

# 面積を計算する
area = tate * yoko   #  take * yoko で 面積を計算、計算した結果をareaに設定

# 面積を出力
print(area)    #
```
