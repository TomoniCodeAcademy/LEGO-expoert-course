Pythonでも同じように、判断して処理を切り替えることができます。条件に応じて実行を変える文をif文と呼びます。上記のブロックをPythonで書き直した例は以下です（イメージをつかんでもらうため、意訳しておりこのままではエラーとなりうごきません。）
センサの値を取得する関数や、モータを動かす関数を除けば、if文のイメージがつかめると思います（きっと）。

```python
if sensor_A.distance() < 5:
    motor.turn('right', 45)
else:
    motor.start('forward')
```

前回説明した比較する式と合わせて実際に動くプログラムの例を説明します。以下の例は年齢を表す変数ageの値が15以上かどうかを判断し、15以上であれば’中学生ですね’と出力する例です。実際に動かしてみください。

```python
age = 15
if age >= 12:
    print('あなたは中学生ですね')
else:
    print('あなたは小学生ですね')
```

2行目のif文の行で、 変数ageと12を比較します。もし変数ageの値が12と同じか、または12よりも大きい場合は、比較した結果がTrueとなり、3行目のprint関数が実行されます。12よりも小さい場合は比較した結果がFalseとなり、5行目のprint関数が実行されます。

補足：TOMO-NIの皆さんは、小学生か中学生しかいないので、上記プログラムで問題ありませんが、TOMO-NIの皆さん以外の人が使うと、間違った答えを返すことになります。

if文の条件をたくさん集めることで、例えば住んでいる住所からどの学校に通っているかを答えさせることもできます。いろんな情報を整理して、大量のif文の形で置き換えることにより、コンピュータはどんどん賢く応答してくれます。
