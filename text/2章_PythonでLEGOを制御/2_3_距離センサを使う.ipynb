# 距離センサの値を取得する

距離センサを使って距離を測るには、距離センサを初期化した後、関数get_distance_cmを使います<br>
file:  SENSOR_distance_00.xxxx

```python
from spike import DistanceSensor


# 距離センサーを初期化します。
distance_sensor = DistanceSensor('A')          # センサはport Aに接続

distance = distance_sensor.get_distance_cm()   # 距離を測る
print(distance)                                # 表示する
```

# 距離センサのLEDを光らせる

距離センサにはLEDが付いていて光らせることができます。LEDを光らせるには関数light_up_allを使います。

![Image Description](https://drive.google.com/file/d/1Bgv0fARmVZGGrdF-NVsoigEYHct5EI14/view?usp=sharing)
<img src="https://drive.google.com/file/d/1Bgv0fARmVZGGrdF-NVsoigEYHct5EI14/view?usp=sharing" alt="Image Description" width="500">


```python
from spike import DistanceSensor

# 距離センサーを初期化します。
distance_sensor = DistanceSensor('A')          # センサはport Aに接続
distance_sensor.light_up_all(100)
```

light_up_all(0)を実行すると距離センサのLEDが消えます

```python
from spike import DistanceSensor

# 距離センサーを初期化します。
distance_sensor = DistanceSensor('A')        # センサはport Aに接続
distance_sensor.light_up_all(0)
```

LEDをチカチカ光らせるには、light_up_all(100)とlight_up_all(0)を順番に実行します<br>
file: SENSOR_distance_blink.###

```python
from spike import DistanceSensor
import time

# 距離センサーを初期化します。
distance_sensor = DistanceSensor('A')        # センサはport Aに接続
while True:
    distance_sensor.light_up_all(100)
    time.sleep(1)
    distance_sensor.light_up_all(0)
    time.sleep(1)
```
