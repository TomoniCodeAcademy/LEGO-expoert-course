# モーションセンサの値を取得する（ピッチ、ロール、ヨー）

モーションセンサを使うことで、HUBの傾きや回転（ピッチ、ロール、ヨー）の角度を取得することができます。HUBの回転角度を得るには、関数get_yaw_anagleを使います<br>
file:  SENSOR_distance_00.xxxx

```python
from spike import PrimeHub

hub = PrimeHub()

angle = hub.motion_sensor.get_yaw_angle()
print(angle)
```

ヨー角については、現在HUBが向いている状態をヨー角０として設定することができます。関数reset_yaw_angleを用います。ヨー角０に設定することで、プログラム実行時のHUBの向きを基準に回転角を決めることが可能になります。

```python
from spike import PrimeHub

hub = PrimeHub()

hub.motion_sensor.reset_yaw_angle()
angle = hub.motion_sensor.get_yaw_angle()
print('Angle:', angle)

```
