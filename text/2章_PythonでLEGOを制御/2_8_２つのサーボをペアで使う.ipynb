# サーボモータをペアで動かす

サーボモータを２つ組み合わせて動かすには、MotorPairモジュールを使います。<br>
MotorMair

関数は大きく分けて、Move と　Startがあります
Moveは指定した分だけ移動して止まります。
Startは移動を開始した後止めるまで動き続けます。

例えば目的の場所まで〇cm前進させたい場合はMoveを使います。目印や色が見つかるまでその場で回転しながら探す場合はStartを使います。

# サーボモータを２台ペアで使いLEGOカーを直進させる

サーボモータを使う上で、LEGOカーを動かす場合はサーボモータを２つペアで動かします。<br>
以下のプログラムはどこまでもまっすぐ進みます。

```python
from spike import MotorPair

motor_pair = MotorPair('A','B')
motor_pair.start(steering=0, speed=20)
```

モータの回転を止めるには 関数stopを使います。

```python
from spike import Motor

motor_pair = MotorPair('A','B')
motor_pair.stop()
```

下記のプログラムは直進した後、１秒後に止まります

```python
from spike import MotorPair
import time

motor_pair = MotorPair('A','B')

motor_pair.start(steering=0, speed=20)
time.sleep(1)
motor_pair.stop()
```

指定した角度まで回す、指定した回転数を回すこともできます。`

# サーボを２台ペアで使い、LEGOカーを左や右に走らせる

サーボモータを２つペアで使ってLEGOカーを右や左に走らせます<br>
以下のプログラムは右に曲がりながら進みます。
10秒走行した後止まります。

```python
from spike import MotorPair
import time

motor_pair = MotorPair('A','B')

motor_pair.start(steering=30, speed=20)
time.sleep(10)
motor_pair.stop()
```

# LEGOカーをその場で回転させます

LEGOカーをその場で回転するには、move関数で、steering=100と指定します。<br>
サーボを１回転する場合は、以下のようにします。

```python
from spike import MotorPair
import time

motor_pair = MotorPair('A','B')
motor_pair.move(1, unit='rotations', steering=100,speed=20)
```

その場回転で右回りに90度回る

```python
from spike import MotorPair
import time

motor_pair = MotorPair('A','B')
motor_pair.move(90, unit='degrees', steering=100,speed=20)
```

注意点として、１回転はサーボの回転でありLEGOカーの回転ではありません。LEGOカーを１回転したい場合は、タイヤの大きさに合わせて回転数を決める必要があります

# LEGOカーを指定した距離進める

LEGOカーを指定した距離進めるには、move関数を使います<br>
次のプログラムは、30cm直進して止まります

```python
from spike import MotorPair

motor_pair = MotorPair('A','B')

motor_pair.set_motor_rotation(17.6, 'cm')            #  タイヤのサイズを指定します
motor_pair.move(30, unit='cm', steering=0,speed=20)  # 30cmまっすぐ進むようmove関数を実行
```
