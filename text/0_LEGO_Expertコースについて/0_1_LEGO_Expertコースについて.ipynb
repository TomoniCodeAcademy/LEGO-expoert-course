# LEGO Expertコースについて

* LEGO ExpertコースはLEGO Advanceを終了された皆さん向けの、より上位のコースです
* 4月～11月はLEGOを使った競技会に参加します。11月から3月はAIカメラとPythonを使ったLEGOモデルの組み立てを行います。
* LEGOの組み立てを通して、Pythonによるプログラミング、AIの使い方を学ぶことができます
    * AIとは何か、何ができて、何ができないか、どのように活用したらよいのか？　どのように使うのかを学びます
  * AIの基本(データの準備、学習、試験運用、問題点抽出、改善)



# LEGO Expertコースで使うツールについて

LEGO Expertコースでは以下のツールや教材を使います
- Google colab
  - LEGO ExpertコースのテキストはGoogle colabで動くように作られています
  - テキストの説明を読んですぐにPythonを動かして確認することができます
  - Google colabはAIプログラミングでも活用されており、強力な開発環境です
- LEGO と LEGO SPIKE IDE (LEGO SPIKE 開発ツール)
  - これまで皆さんが使ってこられたLEGO SPIKEのプログラミングツールです。LEGO ExpertコースではPython を使います
- AIカメラ
  - LEGO ExpertコースではAIカメラを使います。AIカメラ単体で画像認識ができます(AIカメラに学習させることで、色を判別したり、顔を識別できます)

Google colabではPythonの基本的な動きを勉強します。勉強した結果をLEGOで試します。colabが練習の場で、LEGOが実践の場とご理解ください。
